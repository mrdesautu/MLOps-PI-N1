# Limpieza de Datos

Se genera una columna con el nombre id, sumando la primera letra de la plataforma con el numero de id perteneciente a la columna id. 

```python
amazon['id'] = 'a' + amazon['show_id']
disney['id'] = 'd' + disney['show_id']
hulu['id'] = 'h' + hulu['show_id']
netflix['id'] = 'n' + netflix['show_id']
```

Mediante una funcion for, todas las columnas no numericas, se pasan a minuscula. Con formato String

```python
for column in netflix.columns:
    # Representar en minúsculas sólo si la columna es categórica
    if column in netflixstr:
        netflix[column] = netflix[column].str.lower()

```

```python
for column in hulu.columns:
    if column in hulustr:
        hulu[column] = hulu[column].str.lower()
```

```python
for column in amazon.columns:
    if column in amazonstr:
        amazon[column] = amazon[column].str.lower()
```

```python
for column in disney.columns:
    if column in disneystr:
        disney[column] = disney[column].str.lower()
```

Se rellenan los valores faltantes de la columna rating con G, que indica pelicula apra para todo publico. 

```python
hulu.rating.fillna('G',inplace= True)
amazon.rating.fillna('G',inplace= True)
disney.rating.fillna('G',inplace= True)
netflix.rating.fillna('G',inplace= True)
```

```python
amazon.rating.head() # se verifican los cambios
```

Mediante funcion to_datetime, se modifico el formato de las dechas a el siguiente:`AAAA-mm-dd`

```python
amazon['date_added'] = pd.to_datetime(amazon['date_added'], infer_datetime_format=True, errors='coerce')
amazon['date_added'].head()
```

```python
disney['date_added'] = pd.to_datetime(disney['date_added'], infer_datetime_format=True, errors='coerce')
disney['date_added'].head()
```

```python
hulu['date_added'] = pd.to_datetime(hulu['date_added'], infer_datetime_format=True, errors='coerce')
hulu['date_added'].head() #se verifican los cambios
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'], infer_datetime_format=True, errors='coerce')
netflix['date_added'].tail(10) #se verifican los cambios
```

El campo duration debe convertirse en dos campos: **duration_int** y **duration_type**. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)


```python
#se generan 2 columnas nuevas, utilizando la funcion split
amazon['duration_int'] = amazon.duration.str.split(' ', expand = True)[0]
amazon['duration_type'] = amazon.duration.str.split(' ', expand = True)[1]
amazon.info()
```

```python
#se unifican las variables seasons y Season a season

amazon["duration_type"]= amazon["duration_type"].replace("Seasons", "season")
amazon["duration_type"]= amazon["duration_type"].replace("Season", "season")
amazon[['duration_int', 'duration_type']].head(20)
```

```python
disney['duration_int'] = disney.duration.str.split(' ', expand = True)[0]
disney['duration_type'] = disney.duration.str.split(' ', expand = True)[1]

disney = disney[disney.columns[:10].tolist() + ['duration_int', 'duration_type'] + disney.columns[10:-2].tolist()]

disney.info()
```

```python
disney["duration_type"]= disney["duration_type"].replace("Seasons", "season")
disney["duration_type"]= disney["duration_type"].replace("Season", "season")
disney[['duration_int', 'duration_type']].head()
```

```python
hulu['duration_int'] = hulu.duration.str.split(' ', expand = True)[0]
hulu['duration_type'] = hulu.duration.str.split(' ', expand = True)[1]
hulu = hulu[hulu.columns[:10].tolist() + ['duration_int', 'duration_type'] + hulu.columns[10:-2].tolist()]
hulu.info()
```

```python
hulu["duration_type"]= hulu["duration_type"].replace("Seasons", "season")
hulu["duration_type"]= hulu["duration_type"].replace("Season", "season")
hulu[['duration_int', 'duration_type']].head()
```

```python
netflix['duration_int'] = netflix.duration.str.split(' ', expand = True)[0]
netflix['duration_type'] = netflix.duration.str.split(' ', expand = True)[1]
netflix = netflix[netflix.columns[:10].tolist() + ['duration_int', 'duration_type'] + netflix.columns[10:-2].tolist()]
netflix.info()
```

```python
netflix["duration_type"]= netflix["duration_type"].replace("Seasons", "season")
netflix["duration_type"]= netflix["duration_type"].replace("Season", "season")
netflix[['duration_int', 'duration_type']].head()
```

```python
#Creo una nueva variable, que contiene los 4 datasets unificados:
df_completo = pd.concat([netflix,disney,hulu,amazon])
df_completo.describe() #se verifica que se hayan realizado los cambios. 
```

```python
# Exportamos el dataframe para ser utilizado por la API
df_completo.to_csv("streaming_database.csv", index=False)
```
